```python
# Reading a json file
import json
import pandas as pd

with open('/content/FPH-symptoms-and-conditions-1.json', 'r') as f:
  data = json.load(f)

df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data['data'])
print(df.head())
```

```python
# Extracting only the condition column now
condition_column = df['condition']
condition_column.head()
```

```python
# Importing csv file
df_csv = pd.read_csv('/content/nodes.csv')
print(df_csv.head())
```

```python
# Checking for similar codes
def is_similar(str1, str2):
    # Convert to lowercase for case-insensitive comparison
    str1_lower = str(str1).lower()
    str2_lower = str(str2).lower()

    # Check if one string is a substring of the other
    return str2_lower in str1_lower or str1_lower in str2_lower

# Get the list of node_names from the CSV DataFrame
node_names = df_csv['node_name'].tolist()

# Check if each condition from the condition_column has a similar match in node_names
conditions_present = condition_column.apply(lambda condition: any(is_similar(condition, node_name) for node_name in node_names))

# Create a DataFrame showing which conditions have a similar node_name
condition_check_df = pd.DataFrame({'condition': condition_column, 'has_similar_node_name': conditions_present})

# COunt of false ones
false_count = condition_check_df[condition_check_df['has_similar_node_name'] == False].shape[0]
print(f"Number of conditions without a similar node_name: {false_count}")
```

```python
# Percentage of has similar node is false
percentage_false = (false_count / len(condition_column)) * 100
print(f"Percentage of conditions without a similar node_name: {percentage_false:.2f}%")
```
